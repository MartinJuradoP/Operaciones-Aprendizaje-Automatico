### Formato en tipo de datos

```python
# Convertir 'gender' a categórico
df['gender'] = df['gender'].astype('category')

# Convertir 'C_api' a categórico
df['C_api'] = df['C_api'].astype('category')

# Convertir las columnas de fecha a datetime
df['firstDay'] = pd.to_datetime(df['firstDay'], format='%Y%m%d%H%M%S')
df['lastDay'] = pd.to_datetime(df['lastDay'], format='%Y%m%d%H%M%S')

# Muestra el DataFrame formateado
print(df.dtypes)
```

### Validación de valores Nulos de las columnas

```python
# Calcular el porcentaje de valores nulos por columna
null_percentage = df.isnull().sum() / len(df) * 100

# Convertir a DataFrame para mejor visualización (opcional)
null_percentage_df = null_percentage.reset_index()
null_percentage_df.columns = ['Column', 'Null Percentage']

# Mostrar resultados
print(null_percentage_df)
```
